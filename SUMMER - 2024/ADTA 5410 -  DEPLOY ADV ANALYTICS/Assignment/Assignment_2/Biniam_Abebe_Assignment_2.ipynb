```python
import pandas as pd
```

```python
#A. Download the “Institutional Characteristics (HD) 2022” data from the IPEDS website. and save it as a xlsx file named college_adm_2022”.

#B.Use the pd.read_xlsx(…., index_col="…") function to read the data into Python. Call the loaded data college. Set the INSTNM as index, using index_col="…". Make sure that you have the directory set to the correct location for the data.
college = pd.read_excel('college_Directory_information_2022.xlsx',index_col="INSTNM")
```

```python
#c.	Check the first five rows of the data.
college.head()
```

```python
college.shape
```

```python
#d.	How many rows are in this data? How many columns?
print("Number of rows: ", college.shape[0]+1) #+1 to account for the header
print("Number of columns: ", college.shape[1])
```

```python
#e.	Delete all the columns except the ones given below. Print the first 5 rows. 
#1	UNITID	Unit ID for institution
#2	INSTNM	Institution name
#3	CITY	City
#4	CONTROL	Control: (1) Public (2) Private non-profit (3) private for profit
#5	HLOFFER	Highest level of offering
#6	STABBR	States

college = college[['UNITID','CITY','CONTROL','HLOFFER','STABBR']]

#Print the first 5 rows. 
college.head()
```

```python
#Delete all the rows(universities/colleges) which do not have at least 4 “NULL” columns, using .dropna(axis=0,thresh=4)  
# function and call the new dataframe filtered_college. Print the first 5 rows of filtered_college.
threshhold=4
filtered_college = college.dropna(axis=0,thresh=threshhold) #Drop rows with at least 4 NULL values

#Print the first 5 rows.
college.head()
```

```python
#How many rows and columns does filtered_college have?
print("Number of rows: ", filtered_college.shape[0]+1) #+1 to account for the header
print("Number of columns: ", filtered_college.shape[1])
```

```python
#h.	Use the describe() method of to produce a numerical summary of the variables in filtered_college data set.
filtered_college.describe()
```

```python
#i.	This part has two steps
#i.	Delete all the rows which has HLOFFER= -3 
filtered_college = filtered_college[filtered_college['HLOFFER'] !=-3]
```

```python
import matplotlib.pyplot as plt

# i. Delete all the rows which have HLOFFER = -3
filtered_college_no_negative = filtered_college[filtered_college['HLOFFER'] != -3]

# Produce a scatterplot matrix of the columns ['CONTROL', 'HLOFFER']
plt.figure(figsize=(10, 6))
plt.scatter(filtered_college_no_negative['CONTROL'], filtered_college_no_negative['HLOFFER'])
plt.xlabel('CONTROL')
plt.ylabel('HLOFFER')
plt.title('Scatterplot of CONTROL vs HLOFFER')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
#j.	Find an appropriate visualization to show how the control of universities (CONTROL) is distributed across states (STABBR).
plt.figure(figsize=(15, 8)) #Set the figure size
sns.countplot(data=filtered_college_no_negative, x='STABBR', hue='CONTROL') #Create a countplot
plt.xlabel('State') #Set the x-axis label
plt.ylabel('Count of Universities') #Set the y-axis label
plt.title('Distribution of University Control across States') #Set the title
plt.legend(title='Control') #Set the legend title
plt.xticks(rotation=90) #Rotate the x-axis labels
plt.show() 
```

```python
#2.	(BONUS- Up to 10% additional) Create your own question and provide the answer. 

#1: Distribution of the highest level of offering (HLOFFER) among public universities

# Filter for public universities (CONTROL = 1)
public_universities = filtered_college_no_negative[filtered_college_no_negative['CONTROL'] == 1]

# Create a histogram for HLOFFER among public universities
plt.figure(figsize=(10, 6))
plt.hist(public_universities['HLOFFER'], bins=10, edgecolor='black')
plt.xlabel('Highest Level of Offering (HLOFFER)')
plt.ylabel('Frequency')
plt.title('Distribution of Highest Level of Offering among Public Universities')
plt.grid(True)
plt.show()
```

```python
#2.	(BONUS- Up to 10% additional) Create your own question and provide the answer. 

# Analyze top 10 states for public universities
# Group by state and count the number of public universities in each state
public_universities_by_state = public_universities.groupby('STABBR').size().reset_index(name='count')

# Sort the states by the count of public universities in descending order
top_states_public_universities = public_universities_by_state.sort_values(by='count', ascending=False).head(10)

# Plot the top states for public universities
plt.figure(figsize=(12, 8))
plt.bar(top_states_public_universities['STABBR'], top_states_public_universities['count'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Number of Public Universities')
plt.title('Top 10 States with Most Public Universities')
plt.grid(axis='y')
# Add labels on top of the bars
for i, count in enumerate(top_states_public_universities['count']):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')
plt.show()
```

```python
# Analyze the distribution of private universities

# Filter for private universities (CONTROL = 2 for private non-profit, 3 for private for-profit)
private_universities = filtered_college_no_negative[filtered_college_no_negative['CONTROL'].isin([2, 3])]

# Group by state and count the number of private universities in each state
private_universities_by_state = private_universities.groupby('STABBR').size().reset_index(name='count').head(10)

# Sort the states by the count of private universities in descending order
sorted_private_universities_by_state = private_universities_by_state.sort_values(by='count', ascending=False)

# Plot the sorted distribution of private universities across states with labels
plt.figure(figsize=(15, 8))
bars = plt.bar(sorted_private_universities_by_state['STABBR'], sorted_private_universities_by_state['count'], color='orange')
plt.xlabel('State')
plt.ylabel('Number of Private Universities')
plt.title('Distribution of Private Universities across States (Sorted)')
plt.xticks(rotation=90)
plt.grid(axis='y')

# Add labels on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, '%d' % int(height), ha='center', va='bottom')

plt.show()
```
