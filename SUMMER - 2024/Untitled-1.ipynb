```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.
netflix_df_90s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
duration = netflix_df_90s['duration'].mode()[0]

```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python

# Create a directed graph
G = nx.DiGraph()

# Define the layers and nodes
layers = {
    "Input Layer (4 Neurons)": ["Input 1", "Input 2", "Input 3", "Input 4"],
    "Hidden Layer 1 (8 Neurons, ReLU)": ["H1_1", "H1_2", "H1_3", "H1_4", "H1_5", "H1_6", "H1_7", "H1_8"],
    "Hidden Layer 2 (8 Neurons, ReLU)": ["H2_1", "H2_2", "H2_3", "H2_4", "H2_5", "H2_6", "H2_7", "H2_8"],
    "Output Layer (3 Neurons, Sigmoid)": ["Output 1", "Output 2", "Output 3"]
}

# Add nodes to the graph
for layer, nodes in layers.items():
    for node in nodes:
        G.add_node(node, layer=layer)

# Connect the nodes
edges = []
for i in range(len(layers["Input Layer (4 Neurons)"])):
    for j in range(len(layers["Hidden Layer 1 (8 Neurons, ReLU)"])):
        edges.append((layers["Input Layer (4 Neurons)"][i], layers["Hidden Layer 1 (8 Neurons, ReLU)"][j]))

for i in range(len(layers["Hidden Layer 1 (8 Neurons, ReLU)"])):
    for j in range(len(layers["Hidden Layer 2 (8 Neurons, ReLU)"])):
        edges.append((layers["Hidden Layer 1 (8 Neurons, ReLU)"][i], layers["Hidden Layer 2 (8 Neurons, ReLU)"][j]))

for i in range(len(layers["Hidden Layer 2 (8 Neurons, ReLU)"])):
    for j in range(len(layers["Output Layer (3 Neurons, Sigmoid)"])):
        edges.append((layers["Hidden Layer 2 (8 Neurons, ReLU)"][i], layers["Output Layer (3 Neurons, Sigmoid)"][j]))

G.add_edges_from(edges)

# Define position of the nodes
pos = {}
x_offset = 0
for layer, nodes in layers.items():
    y_offset = len(nodes) / 2 * 10
    for i, node in enumerate(nodes):
        pos[node] = (x_offset, y_offset - i * 10)
    x_offset += 30

# Draw the nodes
layer_colors = {
    "Input Layer (4 Neurons)": "skyblue",
    "Hidden Layer 1 (8 Neurons, ReLU)": "lightgreen",
    "Hidden Layer 2 (8 Neurons, ReLU)": "lightcoral",
    "Output Layer (3 Neurons, Sigmoid)": "orange"
}
node_colors = [layer_colors[G.nodes[node]['layer']] for node in G.nodes]

nx.draw(G, pos, with_labels=True, node_size=3000, node_color=node_colors, font_size=10, font_weight='bold', arrowsize=20, edge_color='gray')

# Draw the edges
nx.draw_networkx_edges(G, pos, edgelist=edges, arrowstyle='-|>', arrowsize=20, edge_color='gray')

# Display the graph
plt.title('Redesigned MLP Network Diagram')
plt.show()
```
