```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = 'EDA_CapstoneAssignmentData.xlsx'
try:
	data = pd.read_excel(file_path)
except FileNotFoundError:
	print(f"File not found: {file_path}")
	data = pd.DataFrame()

# Display the first few rows of the dataframe to understand its structure
if not data.empty:
	print(data.head())
else:
	print("Dataframe is empty. Please check the file path and try again.")
```

```python
# Create a scatter plot to compare the estimated and actual labor costs
plt.figure(figsize=(12, 8))  # Set the size of the plot
plt.scatter(data['Estimated'], data['Actual'], alpha=0.6, color='blue', label='Data Points')  # Create a scatter plot with blue dots
plt.title('Estimated vs Actual Labor Costs', fontsize=16)  # Set the title of the plot
plt.xlabel('Estimated Labor Costs', fontsize=14)  # Set the x-axis label
plt.ylabel('Actual Labor Costs', fontsize=14)  # Set the y-axis label
plt.grid(True, linestyle='--', alpha=0.7)  # Display gridlines with custom style

# Add a reference line (y=x) to indicate perfect estimation
plt.plot([data['Estimated'].min(), data['Estimated'].max()], 
		 [data['Estimated'].min(), data['Estimated'].max()], 
		 color='red', linestyle='--', linewidth=2, label='Perfect Estimation')

# Add annotations for some points
for i in range(len(data)):
	plt.annotate(data['Product'][i], (data['Estimated'][i], data['Actual'][i]), fontsize=9, alpha=0.7)

plt.legend()  # Display the legend
plt.show()  # Display the plot with the reference line
```
