```python
all_documents = []
```

```python
import warnings
warnings.filterwarnings('ignore', message='.*CropBox missing from /Page.*')
```

```python
for blob in blobs: #
  print(str(blob))

  if blob.name.endswith("/") or not blob.name.endswith(".pdf"):
    continue
  print(f"Loading document: {blob.name}")

  loader = GCSFileLoader(
        project_name=PROJECT_ID,
        bucket=BUCKET_NAME,
        blob=blob.name
    )

  documents_from_blob = loader.load()

  document_name = blob.name.split("/")[-1]

  print(f"Document name: {document_name}")

  doc_source_prefix = f"gs://{BUCKET_NAME}"
  doc_source_suffix = "/".join(blob.name.split("/")[1:])
  print(f"Document source suffix: {doc_source_suffix}")
  doc_source = f"{doc_source_prefix}/{doc_source_suffix}"
  print(f"Document source: {doc_source}")

  for document in documents_from_blob:
    document.metadata["source"] = doc_source
    document.metadata["document_name"] = document_name

    all_documents.extend(documents_from_blob)


print(f"Number of documents: {len(all_documents)}")
```

```python
len(all_documents[:50])
```

```python
# split the documents into Chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
    separators=["\n\n", "\n", ".", "!", "?", ",", " ", ""],
)
```

```python
doc_splits = text_splitter.split_documents(all_documents[10:])

for idx,split in enumerate(doc_splits):
  split.metadata["chunk"] = idx

print(f"Number of splits: {len(doc_splits)}")
```

```python
template = """
              Context: You are an expert financial analyst assistant. Your task is to analyze and synthesize a specific excerpt from an SEC filing document to directly address a user's query. You have deep knowledge of financial terminology, accounting principles, and SEC regulations.

              User Query:
              "What are the Risk Factors of DH Enchantment Inc.?"

              Context:
              ---
              [Insert only the section of the SEC filing document that explicitly discusses Risk Factors for DH Enchantment Inc. Remove any extraneous information or sections unrelated to this specific query.]
              ---

              Instructions:
              1. Focus on the Query: The user is asking for the Risk Factors specific to DH Enchantment Inc.
              2. Analyze the Excerpt: Carefully examine the provided excerpt to identify statements explicitly mentioning Risk Factors for DH Enchantment Inc.
              3. Extract Relevant Information: Isolate sentences or bullet points that directly list or describe these Risk Factors. Disregard any content unrelated to Risk Factors or DH Enchantment Inc.
              4. Synthesize and Respond: Based solely on the extracted information, provide a concise and informative response listing the Risk Factors.
                  * Begin your response with "The identified Risk Factors for DH Enchantment Inc. include:" and then present a bulleted list of the extracted Risk Factors.
                  * If the excerpt lacks relevant information, respond with: "I don't know."

              Response:
             """
```

```python
#from re import VERBOSE
reteriveal_qa = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type = "stuff",
    retriever = retriever,
    return_source_documents = True,
    verbose = True,
    chain_type_kwargs = {
        "prompt": PromptTemplate
        (
                        template = template,
            input_variables = ["context","question"],
        ),
    },
)
```

```python
# enable for troublshooting
reteriveal_qa.combine_documents_chain.verbose = False
reteriveal_qa.combine_documents_chain.llm_chain.verbose = False
reteriveal_qa.combine_documents_chain.llm_chain.llm.verbose = False
```

```python
def formatter(result):
  print(f"questions: {result['query']}")
  print("."*80)
  print(f"Answer: {result['result']}")
  print(f"Sources: {result['source_documents']}")
  if "source_documents" in result.keys():
    for idx,ref in enumerate(result["source_documents"]):
      print("."*80)
      print(f"Reference #: {idx}")
      print("."*80)
      if "score" in ref.metadata:
        print(f"Matching Score: {ref.metadata['score']}")
      if "source" in ref.metadata:
        print(f"Source: {ref.metadata['source']}")
      if "document_name" in ref.metadata:
        print(f"Document Name: {ref.metadata['document_name']}")
      print("."*80)
      print(f"Content: \n{ref.page_content}")
  print("."*80)
  print(f"Response:{wrap(result['result'])}")
  print("."*80)

def wrap(s):
  return "\n".join(textwrap.wrap(s, width=120, break_long_words=False))


def ask(
    query,
    qa=reteriveal_qa,
    k=NUMBER_OF_RESULTS,
    search_distance_threshold=SEARCH_DISTANCE_THRESHOLD,
    filters={}
):
  reteriveal_qa.retriever.search_kwargs['search_distance'] = search_distance_threshold
  reteriveal_qa.retriever.search_kwargs['k'] = k
  reteriveal_qa.retriever.search_kwargs['filters'] = filters
  result = reteriveal_qa({"query":query})
  formatter(result)
```
