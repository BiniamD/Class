```python
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pylab as plt

%matplotlib inline
```

```python
# Load the Boston Housing dataset into a Pandas DataFrame
housing_df=pd.read_csv('BostonHousing.csv')
```

```python
# Display the summary statistics
housing_df.describe()
```

```python
#Draw a scatterplot of the variables
housing_df.plot.scatter(x='NOX', y='INDUS')
plt.title('Relationship between Nitric oxides concentration and Proportion of non-retail business acres per town')

plt.xlabel('Nitric oxides concentration (parts per 10 million)')  # set x-axis label
plt.ylabel('Proportion of non-retail business acres per town')  # set y-axis label

plt.show()
```

```python
corr_matrix=housing_df.corr().round(1)
sns.heatmap(corr_matrix,xticklabels=corr_matrix.columns,yticklabels=corr_matrix.columns,annot=True)
plt.show()
```
