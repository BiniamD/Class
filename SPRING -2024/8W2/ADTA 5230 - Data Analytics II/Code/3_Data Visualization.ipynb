```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# Data 1: Apple stock prices from 2013 to 2023
apple_df=pd.read_excel('Apple.xlsx')
apple_df['Date']=pd.to_datetime(apple_df.Date, format='%d%m%Y')
apple_df.head()
```

```python
# Data 2: Boston House prices from http://lib.stat.cmu.edu/datasets/boston

housing_df=pd.read_csv('BostonHousing.csv')
housing_df.head(10)
```

```python
apple_ts = pd.Series(apple_df.Apple.values, index=apple_df.Date)
apple_ts.plot( legend=False, figsize=[6, 4],color='red')
plt.xlabel('Date')  # set x-axis label
plt.ylabel('Stock Price ($)')  # set y-axis label
plt.title('Apple Stock Prices')
plt.tight_layout()
plt.show()
```

# Boxplot

```python
ax=apple_df.boxplot()
ax.set_ylabel('Prices($)')
plt.title('Apple Stock Prices')
plt.show()
```

```python
apple_df.describe()
```

# Scatter Plot

```python
housing_df.plot.scatter(x='LSTAT', y='MEDV')
plt.title('Relationship between % Lower Status of Population and Median Home Value')

plt.xlabel('% of Lower Status')  # set x-axis label
plt.ylabel('Median House Prices')  # set y-axis label

plt.show()
```

```python
housing_df.plot.scatter(x='CRIM', y='MEDV')
plt.title('Relationship between % Lower Status of Population and Median Home Value')
plt.show()
```

```python
sns.scatterplot(x='LSTAT', y='MEDV',data=housing_df,color='red')
```

 # Bar Chart

```python
ax=housing_df.groupby('CHAS').mean().MEDV.plot(kind='bar',figsize=[5,3])
ax.set_ylabel('Avg. MEDV')
plt.show()
```

```python
ax=housing_df.groupby('CHAS').mean().AGE.plot(kind='bar',figsize=[5,3])
ax.set_ylabel('Avg. Age')
plt.show()
```

```python
housing_df.groupby('CHAS').mean()
```

# Histograms

```python
ax=housing_df.MEDV.hist()
ax.set_xlabel('MEDV')
ax.set_ylabel('count')
plt.show()
```

```python
housing_df['MEDV'].plot(bins=50,kind='hist',color='red')
```

# Boxplot

```python
ax=housing_df.boxplot(column='MEDV', by='CHAS')
ax.set_ylabel('MEDV')
plt.suptitle('')
plt.title('Median House Prices Grouped by Being Near the River')
plt.show()
```

# Correlation

```python
corr=housing_df.corr().round(1)
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,annot=True)
plt.show()
```

```python
# Include information about values
fig, ax = plt.subplots()
fig.set_size_inches(11, 7)
sns.heatmap(corr, annot=True, fmt=".1f", cmap="RdBu", center=0, ax=ax)

plt.show()
```
